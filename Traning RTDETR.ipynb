```python
from ultralytics import RTDETR

model = RTDETR('rtdetr-l.pt')   # l = large, can also use rtdetr-x.pt
model.train(
    data="/content/dataset/data.yaml",
    epochs=100,
    imgsz=640,
    batch=8,
    lr0=0.001,
    iou=0.9,
    device=0   # GPU
)
```

```python

from ultralytics import RTDETR
model = RTDETR('/content/runs/detect/train/weights/best.pt')  # adjust if folder name differs
model.predict(
    source='/content/dataset/valid/images',
    conf=0.2,
    save=True
)
```
